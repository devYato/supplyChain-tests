```python
import pandas as pd
import numpy as np
```

```python
PATH = '/DF/panvel_fifth_train.csv'
```

```python
df = pd.read_csv(PATH, sep=';')
df.head()
```

```python
df_col = df.columns
print(df_col)
```

```python
df = df[['product','ano', 'price', 'date', 'sales']]
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['month'] = df['date'].dt.month
df.head()
print(df)
```

```python
df_x = df.groupby(['product', 'ano','month']).agg(price=('price', np.mean), sales=('sales', 'sum'))
df_x.head()
```

```python
df_x = df_x.reset_index()
```

```python
df_x['date'] = pd.to_datetime(df_x['ano'].astype(str) + '-' + df_x['month'].astype(str), format='%Y-%m')
df_x.head()
```

```python
df_x['date'] = df_x['date'].astype(str)
df_x['date'] = df_x['date'].str.slice(stop=7)
df_x.head()
```

```python
pivoted = df_x.pivot(index='product', columns='date', values='sales')
pivoted.head()
```

```python
print(pivoted.columns)
pivoted[['2021-01', '2021-02', '2021-03', '2021-04', '2021-05', '2021-06',
       '2021-07', '2021-08', '2021-09', '2021-10', '2021-11', '2021-12',
       '2022-01', '2022-02', '2022-03', '2022-04', '2022-05', '2022-06']] = pivoted[['2021-01', '2021-02', '2021-03', '2021-04', '2021-05', '2021-06',
       '2021-07', '2021-08', '2021-09', '2021-10', '2021-11', '2021-12',
       '2022-01', '2022-02', '2022-03', '2022-04', '2022-05', '2022-06']].fillna(0)
pivoted.head()
```

```python
pivoted.reset_index(inplace=True)
```

```python
pivoted.head()
```

```python
pivoted['price'] = pd.Series([0]* 945860)
print(pivoted.columns)
```

```python
price_dict = dict(zip(df_x['product'], df_x['price']))
```

```python
pivoted['price'] = pivoted['product'].map(price_dict)
pivoted.head()
```
